```python
#Program to Perform Z Test on Laplace Distribution
from __future__ import print_function
__author__= 'Praveen Ashok Kumar'

import pylab as pl
# this package sets up pretty plots
import seaborn
import scipy.stats as stats
import numpy as np
    
%pylab inline
```

## We create a Laplace Distribution of Mean 100 and 300 Samples for 1000 Distributions Considering a pvalue of 0.05

#### Generating Laplace Distribution

```python
np.random.seed(666)
df = 100;
mysize=1000;
```

```python
lp=np.random.laplace(loc = df, scale = 1.0, size=mysize)
```

```python
pl.figure(figsize=(8,8))
pl.hist(lp,bins=30,stacked=True)
pl.xlabel('Sample Mean', fontsize = 18)
pl.ylabel('No. of Samples', fontsize = 18)
pl.title("Laplace Distribution", fontsize = 18)
pl.show()
```

### Generating 300 Samples of Laplace Distribution with the same Mean & Std. Deviation

```python
lapSamples = []
lapSamplesMean = np.zeros(1000)
lapSamplesZ = np.zeros(1000)
for i in range(1000):
    lapSample = np.random.choice(lp, 100)
    lapSamples.append(lapSample) 
    lapSamplesMean[i] = lapSample.mean()
    Z = (lapSample.mean() - lp.mean()) / (lp.std() / 10.)     #Z-test
    lapSamplesZ[i] = Z
```

```python
plt.hist(lapSamples[i-1], bins = 30, stacked=True)
plt.ylabel('n', fontsize = 14)
plt.xlabel('x', fontsize = 14)
plt.title('Sample %d'%(i), fontsize = 14)
pl.title('Printing a Random Sample')
```

```python
#plot the distribution of z statistics
plt.figure(figsize=(10,6))
plt.hist(lapSamplesZ, bins = 30, stacked=True)
plt.ylabel('N', fontsize = 18)
plt.xlabel('Z statistics', fontsize = 18)
plt.title('The Z statistics of sample means', fontsize = 22)
```

```python
print (stats.kstest(lapSamplesZ, 'norm'))
```

```python
#plot the distribution of z statistics
Gaus = np.random.normal(0, 1, 1000)
plt.figure(figsize=(10,6))
plt.hist(lapSamplesZ, bins = 30, stacked=True)
plt.hist(Gaus, bins = 30, stacked=True)
plt.ylabel('N', fontsize = 18)
plt.xlabel('Z statistics', fontsize = 18)
plt.title('Normal Fit to Z Statistics', fontsize = 22)
```

```python
print (stats.kstest(lapSamplesZ, 'norm'))
```

### Null Hypothesis: Laplace Distribution Follows Gaussian Distribution 
### Since the P Value is 0.19 and it is greater than the assumed p value of 0.05 for the hypothesis, the Null Hypothesis cannot be rejected.

### Therefore, Laplace Distribution considering a 1000 Samples follows a Gaussian Distribution.

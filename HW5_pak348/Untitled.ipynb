```python
#Program to Perform Z Test on Laplace Distribution
from __future__ import print_function
__author__= 'Praveen Ashok Kumar'

import pylab as pl
# this package sets up pretty plots
import seaborn
import scipy.stats as stats
import numpy as np
    
%pylab inline
```

#### Generating Laplace Distribution

```python
np.random.seed(666)
df = 100;
mysize=500;
```

```python
lp=np.random.laplace(loc = df, scale = 1.0, size=mysize)
```

```python
pl.figure(figsize=(8,8))
pl.hist(lp,bins=30,stacked=True)
pl.xlabel('Sample Mean', fontsize = 18)
pl.ylabel('No. of Samples', fontsize = 18)
pl.title("Laplace Distribution", fontsize = 18)
pl.show()
```

```python
lp.mean()
```

```python
lp.std()
```

### Generating 100 Samples of Laplace Distribution with the same Mean & Std. Deviation

```python
lapSamples = []
lapSamplesMean = np.zeros(1000)
lapSamplesZ = np.zeros(1000)
for i in range(1000):
    lapSample = np.random.choice(lp, 100)
    lapSamples.append(lapSample) 
    lapSamplesMean[i] = lapSample.mean()
    Z = (lapSample.mean() - lp.mean()) / (lp.std() / 10.)     #Z-test
    lapSamplesZ[i] = Z
```

```python
plt.figure(figsize = (18,18))
for i in range(1,10):
    plt.subplot(330+i)
    plt.hist(lapSamples[i-1], bins = 30, color='mediumseagreen', alpha=0.7)
    plt.ylabel('n', fontsize = 14)
    plt.xlabel('x', fontsize = 14)
    plt.title('Sample %d'%(i), fontsize = 14)
```

```python
#plot the distribution of z statistics
plt.figure(figsize=(10,6))
plt.hist(lapSamplesZ, bins = 30, color='mediumseagreen', alpha=0.7)
plt.ylabel('N', fontsize = 18)
plt.xlabel('Z statistics', fontsize = 18)
plt.title('The Z statistics of sample means', fontsize = 22)
```

```python
print ("Z statistics on normal", stats.kstest(lapSamplesZ, 'norm'))
```
